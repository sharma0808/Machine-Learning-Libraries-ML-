# Pandas:

### Pandas deal with three types of data structures:
    Series
    Dataframe
    Panel

###### Pandas Series :
- A Pandas Series is like a column in a table.
- It is a one-dimensional array holding data of any type.

###### Pandas DataFrame : 
Pandas DataFrame is two-dimensional size-mutable, potentially heterogeneous tabular data structure with labeled axes (rows and columns). A Data frame is a two-dimensional data structure, i.e., data is aligned in a tabular fashion in rows and columns. Pandas DataFrame consists of three principal components, the data, rows, and columns.

# Pandas dataframe and indexing

##### Note:  we have multiple columns we use pandas dataframe

```python
list = [1,2,3,4]
labels = ['cricket','football','basketball','golf']
sports1 = pd.Series(list,index=labels)
sports1
```

```python
sports1 = pd.Series([1,2,3,4],index=['cricket','football','basketball','golf'])
sports1
```

```python
sports1['basketball']
```

```python
sports2 = pd.Series([11,13,12,4],index=['cricket','football','baseball','golf'])
sports2
```

```python
sports2['baseball']
```

```python
sports1 + sports2 ## Whenever you add/perform_any_operation two Tables/two DataFrames, it shouled be present in both side otherwise it will not return values(return NaN)
                  ## NaN , standing for not a number, is a numeric data type used to represent any value that is undefined or unpresentable. ... The square root of a negative number is an imaginary number that cannot be represented as a real number, so, it is represented by NaN.
```

```python
import numpy as np
import pandas as pd
df1=pd.DataFrame(np.random.rand(8,5),index='A B C D E F G H'.split(),columns='Score1 Score2 Score3 Score4 Score5'.split())
df1
```

```python
df1['Score1']  ### Whenever you slicing the data - use [] 
```

```python
df1[['Score1','Score3','Score5']] ## if you want multiple columns i create a list
```

```python
df1['Score6'] = df1['Score1'] + df1['Score2'] 
df1
```

```python
df1.drop('Score 5',axis=1) ## whenever you write the function, use ().
                          ## if you do not write axis=1, it will go first at check only in the row and say that there is no "Score5" in that.
                          ## whenever you delete a row you write "axis=0" ------( by Default row)
                          ## whenever you delete a column you write "axis=1"
```

```python
df1['Score7'] = df1['Score1'] + df1['Score2'] 
df1
```

```python
df1.drop(['Score5','Score6'],axis=1) ## drop multiple columns
```

```python
df1.drop('E',axis=0)
```

```python
df1.drop(['D','F']) ## Drop multiple Rows
```
